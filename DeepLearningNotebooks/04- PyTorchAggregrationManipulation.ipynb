```python
# manipulating tensor : reshaping, stacking, squeezing, unsqueezing
```

```python
x = torch.arange(1,10)
x
```

```python
print(x.shape)
print(x.ndim)
```

```python
# Elimizdeki verinin boyutunu değiştirmek için kullanılır. -> reshape()
x_reshaped = x.reshape(1,9)
print(x_reshaped)
print(x_reshaped.shape)
print(x_reshaped.ndim)
```

```python
# view
```

```python
x_reshaped = x.reshape(9,1)
x_reshaped
```

```python
x_reshaped[0] = 20
```

```python
x_reshaped
```

```python
x_view = x.view(9,1)
x_view
```

```python
x_view[0] = 10
x_view
```
