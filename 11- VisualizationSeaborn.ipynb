```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Veriler/athlete_events.csv")
```

```python
data.describe()
```

```python
data.head()
```

```python
data.isnull()
```

```python
data.info()
```

```python
plt.scatter("Height","Weight",data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")
plt.show()
```

```python
sns.set_style("dark") # darkgrid, whitegrid, white, dark
sns.scatterplot(x="Height",y="Weight",data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid") # darkgrid, whitegrid, white, dark
sns.scatterplot(x="Height",y="Weight",hue="Sex",data=data) 
# hue ise daha çok kategorik verilerde kullanılır.
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
data.columns
```

```python
data["Medal"].unique()
```

```python
sns.set_style("whitegrid") 
sns.scatterplot(x="Height",y="Weight",hue="Sex",style="Medal",size="Age",data=data) 
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height",y="Weight",data=data,hue="Sex")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight (Lineplot)")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height",hue="Sex",data=data)
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height",hue="Sex",data=data,kind="kde")
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal",y="Height",hue="Sex",data=data)
plt.title("Medals by Height")
plt.show()
```

```python
sns.catplot(x="Medal",y="Height",data=data,hue="Sex",col="Season")
plt.show()
```
