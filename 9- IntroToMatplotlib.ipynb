```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list= [10,20,30,40,50,60,70,75,60,45]
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list,weight_list,"g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
# plt.savefig() -> Bu da bulunduğu dizine kaydeder.
```

```python
#numpy
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
np_array_1 = np.linspace(0,10,20)
np_array_2 = np_array_1 ** 3
```

```python
plt.plot(np_array_1,np_array_2,"b*-")
```

```python
# yan yana 2 grafik görme
```

```python
plt.subplot(1,2,1)
plt.plot(np_array_1,np_array_2,"y*-")
plt.subplot(1,2,2)
plt.plot(np_array_2,np_array_2,"g--")
plt.show()
```

```python
# Kendi figürlerimizi değiştirme oluşturma.
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.5,0.7])
figure_axes.plot(np_array_1,np_array_2,"g")
figure_axes.set_xlabel("X axis")
figure_axes.set_ylabel("Y axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_figure = plt.figure(dpi=200)
new_axes = new_figure.add_axes([0.1,0.1,0.5,0.5])
new_axes.plot(np_array_1,np_array_1**2,label="Numpy array ** 2")
new_axes.plot(np_array_1,np_array_1**3,label="Numpy array ** 3")
new_axes.legend(loc=6)
plt.show()
```
