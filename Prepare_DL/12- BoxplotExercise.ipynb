```python
data = np.array([5,7,9,13,16,17,19,22,28,34,35,40,46,49,50,57,103])
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box plot")
plt.ylabel("Y Values")
#plt.grid(True)
plt.show()
```

```python
df = sns.load_dataset("titanic")
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
sns.boxplot(x='class',y='age',data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x='class',y='fare',data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.tight_layout()
plt.show()
```
