```python
import numpy as np
```

#### Numpy Dizileri

```python
my_list = [10,20,30,40]
type(my_list)
```

```python
# np.array() ile dizi oluşturuyoruz.
my_list = np.array(my_list)
```

```python
# Yeni type ndarray oluyor.
type(my_list)
```

```python
my_list.max() # maximum değeri getirir.
```

```python
print(np.ones(5)) # içerisinde 5 tane 1 olan dizi oluşturur.
print(np.zeros(5)) # içerisinde 5 tane 0 olan dizi oluşturur.
print(np.random.random(5)) # içerisinde 5 tane random sayı olan dizi oluşturur.
```

```python
my_list_1 = [1,2]
my_list_2 = [3,4]
my_list_1 + my_list_2
```

```python
# Numpy'de diziler toplanırken normaldeki gibi yan yana yazılmaz. Değerleri toplar.
yeni_toplam = np.array(my_list_1) + np.array(my_list_2)
yeni_toplam
```

#### Numpy Indexler

```python
# arange & indexing
```

```python
np.arange(0,10,2)
```

```python
my_np_array = np.arange(0,10)
my_np_array
```

```python
# normal diziler gibi indexing yapısı vardır.
print(my_np_array[0])
print(my_np_array[-1])
print(my_np_array[3:5])
```

```python
np.random.randn(4,4) # 4x4 bir matris oluşturur. Sayılar rastgeledir.
```

```python
np.random.randint(1,300,5) # 1 ile 300 arasında 5 tane rastgele değer üretir. 
```
