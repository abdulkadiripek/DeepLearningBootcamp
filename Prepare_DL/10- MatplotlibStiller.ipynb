```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1 ** 2
```

```python
my_fig, my_axes = plt.subplots()
```

```python
type(my_axes)
```

```python
type(my_fig)
```

```python
my_fig,my_axes = plt.subplots()
my_axes.plot(data1,data2,color="magenta",alpha=0.5)
my_axes.plot(data2,data1,color="cyan")
plt.show()
```

```python
(new_fig,new_axes) = plt.subplots()
new_axes.plot(data1,data1+2,color="Blue",linewidth=4.0)
new_axes.plot(data1,data1+3,color="Magenta",linewidth=1.0)
new_axes.plot(data1,data1+4,color="Magenta",linestyle="-.")
new_axes.plot(data1,data1+5,color="Magenta",linestyle=":")
new_axes.plot(data1,data1+6,color="Green",linestyle="--",marker="o",markersize=5,markerfacecolor="red")
new_axes.plot(data1,data1+7,color="Magenta",linestyle="--",marker="h",markersize=5)
new_axes.plot(data1,data1+8,color="Magenta",linestyle="--",marker="+",markersize=6,markerfacecolor="red")
plt.show()
```

```python
plt.scatter(data1,data2,marker="h",color="Black")
plt.show()
# Scatter nokta nokta grafik çizer.
```
