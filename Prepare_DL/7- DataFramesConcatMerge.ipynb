```python
# merge
```

```python
df_merge1 = pd.read_csv('Veriler/7-merge_data1.csv')
```

```python
df_merge1
```

```python
df_merge2 = pd.read_csv('Veriler/7-merge_data2.csv')
```

```python
df_merge2
```

```python
# Merge - Inner Join
```

```python
merged_df = pd.merge(left=df_merge1,right=df_merge2,how="inner",on="Employee_ID")
```

```python
merged_df
```

```python
# Merge - outer join
```

```python
df_merged_outer = pd.merge(df_merge1,df_merge2,on="Employee_ID",how="outer")
```

```python
df_merged_outer
```

```python
# Merge - left join
```

```python
df_merged_left = pd.merge(df_merge1,df_merge2,how="left",on="Employee_ID")
```

```python
df_merged_left
```

```python
# merge - right join
```

```python
df_merge_right = pd.merge(df_merge1,df_merge2,on="Employee_ID",how="right")
```

```python
df_merge_right
```
