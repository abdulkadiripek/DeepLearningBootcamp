```python
import numpy as np
import pandas as pd
```

### Pandas Series

```python
grades = {"Kadir":50,"Ozge":60,"Atil": 80}
```

```python
pd.Series(grades)
```

```python
names = ["Kadir", "Ozge", "Atil"]
grades = [50,60,70]
pd.Series(data=grades,index=names,name="Ogrenci Notlari")
```

```python
# with numpy
numpy_array = np.array([10,20,30,40,50])
pd.Series(numpy_array)
```

```python
# arithmetic
contest_result = pd.Series(data=[10,5,100],index=["Kadir","Ozge","Atil"])
contest_result2= pd.Series(data=[50,100,4],index=["Kadir","Ozge","Atil"])
contest_result
```

```python
contest_result["Ozge"]
```

```python
final_result = contest_result + contest_result2
final_result
```

```python
different_series = pd.Series([10,20,30,50,70],["a","b","c","d","e"])
different_series
```

```python
different_series2 = pd.Series([5,4,3,2,1],["a","c","b","f","g"])
different_series + different_series2
```
