```python
from torch import nn  # neural network demek (nn)
```

```python
class SimpleLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1,dtype=torch.float),requires_grad=True)
        self.bias = nn.Parameter(torch.randn(1,dtype=torch.float),requires_grad=True) 

    def forward(self,x:torch.Tensor) -> torch.Tensor:
        return self.weights * x + self.bias        
```

```python
torch.manual_seed(42)
model_0 = SimpleLinearRegressionModel()
model_0 = torch.compile(model_0) # For the optimization.
```

```python
list(model_0.parameters())
```

```python
model_0.state_dict() # içinde bulunduğumuz güncel durumun sözlüğü.
```

```python
with torch.inference_mode():
    y_pred = model_0(x_test)
```
