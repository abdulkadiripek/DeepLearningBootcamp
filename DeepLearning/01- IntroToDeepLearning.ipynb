```python
from torch import nn  # neural network demek (nn)
```

```python
class SimpleLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.weights = nn.Parameter(torch.randn(1,dtype=torch.float),requires_grad=True)
        self.bias = nn.Parameter(torch.randn(1,dtype=torch.float),requires_grad=True) 

    def forward(self,x:torch.Tensor) -> torch.Tensor:
        return self.weights * x + self.bias        
```

```python
torch.manual_seed(42)
model_0 = SimpleLinearRegressionModel()
model_0 = torch.compile(model_0) # For the optimization.
```

```python
list(model_0.parameters())
```

```python
model_0.state_dict() # içinde bulunduğumuz güncel durumun sözlüğü.
```

```python
with torch.inference_mode():
    y_pred = model_0(x_test)
```

```python
print(len(x_test))
print(len(y_pred))
print(len(y_test))
```

```python
plt.scatter(x=x_train,y=y_train,c="blue", label = "Training Data", s=6)
plt.scatter(x_test,y_test,c="green",label="Test Data",s=7)
plt.scatter(x_test,y_pred,c="red",label="Test Data",s=10)
plt.show()
```

```python
# loss ve optimizer olmadan olmaz.
```

```python
loss_func = nn.MSELoss()
# loss_func=nn.L1Loss() # MAE
optimizer = torch.optim.SGD(params = model_0.parameters(), lr= 0.001)

torch.manual_seed(42)
epochs = 200
```

```python
train_loss_values = []
test_loss_values = []
epoch_count = []
```

```python
for epoch in range(epochs):
    # modeli train moduna sokar.
    model_0.train()

    y_pred = model_0(x_train)
    loss = loss_func(y_pred,y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    model_0.eval()

    with torch.inference_mode():
        test_pred = model_0(x_test)
        test_loss = loss_func(test_pred,y_test) 

        if epoch % 5 == 0:
            epoch_count.append(epoch)
            train_loss_values.append(loss.detach().numpy())
            test_loss_values.append(test_loss.detach().numpy())
            print(f"Epoch: {epoch}, Train Loss: {loss}, Test Loss: {test_loss}")
```

```python
plt.plot(epoch_count,train_loss_values,label="Train Loss")
plt.plot(epoch_count,test_loss_values, label="Test Loss" )
plt.ylabel("Loss")
plt.xlabel("Epochs")
plt.show()
```
